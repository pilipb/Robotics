```python
import seaborn as sns
import pandas as pd

# Here, these are distance between the
# 3Pi+ center and the line center.
pwm_30 = [68, 69, 67.5, 67.3, 67.7] # cm in 5 seconds
pwm_45 = [112, 111.7, 112.9, 112.5, 113]

# normalizing the data using the pwm value
pwm_30 = [x / 30 for x in pwm_30]
pwm_45 = [x / 45 for x in pwm_45]

pwm_30_dataframe = pd.DataFrame({'20':pwm_30})
pwm_45_dataframe = pd.DataFrame({'40':pwm_45})
# pwm_60_dataframe = pd.DataFrame({'60':pwm_60})

# We then add the two dataframes together.
combined = pd.concat([ pwm_30_dataframe, pwm_45_dataframe], axis=0)
# combined = pd.concat([ pwm_20_dataframe, pwm_40_dataframe, pwm_60_dataframe], axis=0)

# We can now plot both datasets at the same time.
# Note, data=combined!
plot = sns.boxplot(data=combined)
plot.set(xlabel ="PWM Value", ylabel = "Travel Error Norm (cm)", title ='My Results')
```
